# Sentiment Text Analysis #

The dataset is taken from https://www.kaggle.com/praveengovi/emotions-dataset-for-nlp.
This is binary classification analysis, where we have 'positive' and 'negative' results.

```python
import pandas as pd
import numpy as np
from PIL import Image 
import matplotlib.pyplot as plt 
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn import metrics 
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve
#nltk.download('wordnet')
#nltk.download('punkt')
```

First we look at given datasets. We see that data is a bit imbalanced: some classes are more frequent than others.

```python
df_train = pd.read_csv('train.txt', delimiter=';', names=['text', 'emotions'])
df_val = pd.read_csv('val.txt', delimiter=';', names=['text', 'emotions'])
df_test = pd.read_csv('test.txt', delimiter=';', names=['text', 'emotions'])
data = {'train': df_train, 'val': df_val, 'test': df_test}
bar, ax = plt.subplots(1,3, figsize=(30, 10))
for index, temp in enumerate(data):
    sns.countplot(ax = ax[index],x = 'emotions', data = data[temp])
    ax[index].set_title(temp+' Class Frequency', size=14)
    ax[index].set_ylabel('Frequency', size=14)
    ax[index].set_xlabel(temp+' Class', size=14)
plt.show()
```

For binary classification we replace _sadness, anger and fear_ with __negative__ and _love, joy and surprise_ with __positive__ labels.

```python
stop_words = set(stopwords.words('english'))
for index,temp in enumerate(data):
    data[temp] = data[temp].replace('sadness', 'negative')
    data[temp] = data[temp].replace('anger', 'negative')
    data[temp] = data[temp].replace('fear', 'negative')
    data[temp] = data[temp].replace('love', 'positive')
    data[temp] = data[temp].replace('joy', 'positive')
    data[temp] = data[temp].replace('surprise', 'positive')

bar, ax = plt.subplots(1,3, figsize=(30, 10))
for index, temp in enumerate(data):
    sns.countplot(ax = ax[index],x = 'emotions', data = data[temp])
    ax[index].set_title(temp+' Class Frequency', size=14)
    ax[index].set_ylabel('Frequency', size=14)
    ax[index].set_xlabel(temp+' Class', size=14)
plt.show()
```

### Data Preprocessing ###
Here we remove digits and other symbols, make lower case, word tokenization, lemmatization and removing stop words.

```python
for index,temp in enumerate(data):
    #remove digits, punctuation and other symbols with regex
    data[temp]['text'] = [(lambda f: re.sub(r'\d|\W', ' ', text))(text) for text in data[temp]['text']]
    #make lowercase
    data[temp]['text'] = [(lambda f: text.lower())(text) for text in data[temp]['text']]
    #word tokenization
    data[temp]['text'] = [(lambda f: [x for x in nltk.word_tokenize(text)])(text) for text in data[temp]['text']]
    #lemmatizing words
    lemmatizer = WordNetLemmatizer()
    data[temp]['text'] = [(lambda word: [(lambda f: lemmatizer.lemmatize(word, pos='v'))(word) for word in text])(text) for text in data[temp]['text']]
    data[temp]['text'] = [(lambda word: [(lambda f: lemmatizer.lemmatize(word, pos='n'))(word) for word in text])(text) for text in data[temp]['text']]
    #stemming
    #stemmer = PorterStemmer()
    #data['text'] = [(lambda word: [(lambda f: stemmer.stem(word))(word) for word in text])(text) for text in data['text']]
    #remove stop words
    data[temp]['text'] = [(lambda f: [x for x in text if not x in stop_words])(text) for text in data[temp]['text']]
    #back to sentences
    data[temp]['text'] = [(lambda f: ' '.join(text))(text) for text in data[temp]['text']]
```

Here we visualize most frequent words with word clouds for each dataset.

```python
stopwords = set(STOPWORDS)
wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)
# generate the word cloud
wc.generate(str(list(data['train']['text'])))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()

wc.generate(str(list(data['val']['text'])))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()

wc.generate(str(list(data['test']['text'])))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### Training Models ###
Here we train different classification models to compare their performance.
We use TfidfVectorizer() in pipeline so that it will take our texts and convert to vectors of integers, defining appearance of words in each text.

```python
x_train, x_test, y_train, y_test = data['train']['text'], data['val']['text'], data['train']['emotions'], data['val']['emotions']


svm_mod = make_pipeline(TfidfVectorizer(), SVC(kernel='linear', probability=True))
svm_mod.fit(x_train, y_train)
pred_svm = svm_mod.predict(x_test)
print("SVM accuracy ",metrics.accuracy_score(y_test, pred_svm))

nb_model = make_pipeline(TfidfVectorizer(), MultinomialNB())
nb_model.fit(x_train, y_train)
pred_nb = nb_model.predict(x_test)
print('Naive Bayes accuracy', metrics.accuracy_score(y_test, pred_nb))

log_model = make_pipeline(TfidfVectorizer(), LogisticRegression())
log_model.fit(x_train, y_train)
pred_log = log_model.predict(x_test)
print('Logistic Regression accuracy', metrics.accuracy_score(y_test, pred_log))

tree = make_pipeline(TfidfVectorizer(), DecisionTreeClassifier())
tree.fit(x_train,y_train)
tree_pred = tree.predict(x_test)
print('Decision Tree accuracy', metrics.accuracy_score(y_test, tree_pred))
```

For KNN algorithm we vizualize its performance depending on different k, and choose best k for model.

```python
#finding k for knn
#testing knn
acc = []
for k in range(1, 50):
    knn = make_pipeline(TfidfVectorizer(), KNeighborsClassifier(n_neighbors=k))
    knn.fit(x_train, y_train)
    pred_knn = knn.predict(x_test)
    acc.append(metrics.accuracy_score(y_test, pred_knn))
    
plt.figure(figsize=(10,6))
plt.plot(range(1,50),acc,color = 'blue',linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Maximum accuracy:",max(acc),"at K =",acc.index(max(acc))+1)
```

```python
k = acc.index(max(acc))+1
knn = make_pipeline(TfidfVectorizer(), KNeighborsClassifier(n_neighbors=k))
knn.fit(x_train, y_train)
pred_knn = knn.predict(x_test)
print('KNN accuracy', metrics.accuracy_score(y_test, pred_knn))
```

### Comparing Models ###
Accuracy score is not enough, let's look at other metrics.

```python
y_test = y_test.replace('negative', 0)
y_test = y_test.replace('positive', 1)
pred_log = pred_log == 'positive'
pred_log = pred_log.astype(int)
pred_svm = pred_svm == 'positive'
pred_svm = pred_svm.astype(int)
pred_knn = pred_knn == 'positive'
pred_knn = pred_knn.astype(int)
tree_pred = tree_pred == 'positive'
tree_pred = tree_pred.astype(int)
pred_nb = pred_nb == 'positive'
pred_nb = pred_nb.astype(int)

#Precision 
logm_precision = metrics.precision_score(y_test, pred_log)
svc_precision = metrics.precision_score(y_test, pred_svm)
knn_precision = metrics.precision_score(y_test, pred_knn)
dt_precision = metrics.precision_score(y_test, tree_pred)
rf_precision = metrics.precision_score(y_test, pred_nb)

#Recall 
logm_recall = metrics.recall_score(y_test, pred_log)
svc_recall = metrics.recall_score(y_test, pred_svm)
knn_recall = metrics.recall_score(y_test, pred_knn)
dt_recall = metrics.recall_score(y_test, tree_pred)
rf_recall = metrics.recall_score(y_test, pred_nb)

#F1-measure
logm_f1 = metrics.fbeta_score(y_test, pred_log, beta=1.0)
svc_f1 = metrics.fbeta_score(y_test, pred_svm, beta=1.0)
knn_f1 = metrics.fbeta_score(y_test, pred_knn, beta=1.0)
dt_f1 = metrics.fbeta_score(y_test, tree_pred, beta=1.0)
rf_f1 = metrics.fbeta_score(y_test, pred_nb, beta=1.0)

#compare models accross classification metrics
Model_Comparison = pd.DataFrame({
    'Model': ['Logistic Regression', 'Support Vector Machine', 'K-Nearest Neighbor', 
              'Decision Tree', 'Naive Bayes'],
    'Accuracy': [metrics.accuracy_score(y_test, pred_log), metrics.accuracy_score(y_test, pred_svm), metrics.accuracy_score(y_test, pred_knn), metrics.accuracy_score(y_test, tree_pred), metrics.accuracy_score(y_test, pred_nb)],
    'Precision': [logm_precision, svc_precision, knn_precision, dt_precision, rf_precision],
    'Recall': [logm_recall, svc_recall, knn_recall, dt_recall, rf_recall],
    'F1-meas.': [logm_f1, svc_f1, knn_f1, dt_f1, rf_f1]})
Model_Comparison.round(4)
```

Plotting ROC of each model.

```python
def plot_roc(yhat, name):
    # retrieve just the probabilities for the positive class
    pos_probs = yhat[:, 1]

    # plot no skill roc curve
    plt.plot([0, 1], [0, 1], linestyle='--', label='No Skill')

    # calculate roc curve for model
    fpr, tpr, threshold = roc_curve(y_test, pos_probs)

    # plot model roc curve
    plt.plot(fpr, tpr, marker='.', label=name)

    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

    # show the legend
    plt.legend()

    # show the plot
    plt.show()
```

```python
plot_roc(log_model.predict_proba(x_test), 'Logistic')
plot_roc(nb_model.predict_proba(x_test), 'Naive Bayes')
plot_roc(svm_mod.predict_proba(x_test), 'SVM')
plot_roc(tree.predict_proba(x_test), 'Decision Tree')
plot_roc(knn.predict_proba(x_test), 'KNN')
```
